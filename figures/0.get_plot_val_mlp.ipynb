```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_mlp(dataset):
    val_df = pd.read_csv(f'../model/mlp/report_{dataset}.csv')
    for num_layers in val_df['num_layers'].unique():
        val_df_layer = val_df[val_df['num_layers'] == num_layers]
        # Assuming val_df is your original DataFrame
        mean_val_df = val_df_layer.groupby('layer_size')[['val_loss', 'train_loss']].mean().reset_index()

        # Convert hidden_size to categorical (as strings)
        mean_val_df['layer_size'] = mean_val_df['layer_size'].astype(str)

        plt.figure(figsize=(8, 3))

        # Plot val_loss with blue color and solid line
        plt.plot(mean_val_df['layer_size'], mean_val_df['val_loss'], color='black', linestyle='-', label='val_loss')

        # Plot train_loss with black color and dashed line
        plt.plot(mean_val_df['layer_size'], mean_val_df['train_loss'], color='black', linestyle='--', label='train_loss')

        # Set specific ticks for the x-axis (hidden_size as categorical)
        plt.xticks(mean_val_df['layer_size'])

        # Add labels and title
        plt.xlabel('Feature Numbers')
        plt.ylabel('Loss')
        plt.title(f'Validation Loss and Train Loss vs Extracted Feature Numbers: {num_layers} layers')
        plt.legend()
        plt.show()
```

```python
plot_mlp('detailed')
```

```python
plot_mlp('systematic')
```
